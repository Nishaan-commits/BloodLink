```python
import pandas as pd
import matplotlib.pyplot as plt

#load dataset
df = pd.read_csv("../data/raw/transfusion.csv")

#Rename columns
df.columns = ["months_since_last_donation","number_of_donations","total_volume_donated","months_since_first_donation","made_donation"]

#check basic info
print(df.info())

#check for missing values
print(df.isnull().sum())

#View basic stats
print(df.describe())

#Visualize class balance
df["made_donation"].value_counts().plot(kind='bar')
plt.title("Donation Distribution")
plt.xlabel("Donated")
plt.ylabel("Count")
plt.show
```

```python
import seaborn as sns

#Distribution of donations
sns.histplot(df['total_volume_donated'],kde=True)
plt.title('Distribution of Blood donated')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()

```

```python
sns.countplot(x='made_donation',data=df)
plt.title("Target Class Distribution")
plt.show()
```

```python
#Checking for duplicates
df = df.drop_duplicates()
print(df.duplicated().sum())

#Check data types
print(df.dtypes)
```

```python
#Save the cleaned/transformed data
df.to_csv("../data/processed/cleaned_data.csv",index=False)
```
